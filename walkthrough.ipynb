```python
N = torch.zeros((28,28))
```

```python
# remove duplicates and put everything into big string -> should be 26 english letters
# set(''.join(words))
# len(set(''.join(words)))
chars = sorted(set(''.join(words)))
```

```python
# mapping chars to numbers
stoi = { s:i for i,s in enumerate(chars)}
stoi['<S>'] = 26
stoi['<E>'] = 27
stoi
```

```python
for w in words :
    chs = ['<S>'] + list(w) + ['<E>']
    for ch1,ch2 in zip(chs,chs[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        N[ix1,ix2] += 1
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.imshow(N)
```

```python
itos = { i:s for s,i in stoi.items()}
itos
```

```python
plt.figure(figsize=(16,16))
plt.imshow(N,cmap='Blues')
for i in range(28):
    for j in range(28):
        chstr = itos[i] + itos[j]
        plt.text(j,i,chstr,ha="center",va="bottom",color="gray")
        plt.text(j,i,N[i,j].item(),ha="center",va="top",color="gray")
plt.axis('off')
```
